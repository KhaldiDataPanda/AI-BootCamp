```python
import pandas as pd
df = pd.read_csv('Data/ChicagoCensusData.csv')
coms = df['COMMUNITY_AREA_NAME']

list_of_means = []
#print(list_of_means)

for com in coms:
    mean_per_com = df[df['COMMUNITY_AREA_NAME'] == com]['PER_CAPITA_INCOME'].mean()
    list_of_means.append(float(mean_per_com))
    #print(f'{com} : {mean_per_com}')
    #print(list_of_means)
    
#plt.figure(figsize=(20, 10))
#sns.barplot(x=coms, y=list_of_means)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
coms = df_census['COMMUNITY_AREA_NAME'].unique()
# Calculate number of subplots needed
num_communities = len(coms)
communities_per_plot = 5
num_subplots = (num_communities + communities_per_plot - 1) // communities_per_plot

# Create subplots
fig, axes = plt.subplots(num_subplots, 1, figsize=(15, 4 * num_subplots))

# If only one subplot, axes is not an array
if num_subplots == 1:
    axes = [axes]

# Plot each group of 5 communities
for i in range(num_subplots):
    start_idx = i * communities_per_plot
    end_idx = min(start_idx + communities_per_plot, num_communities)
    
    coms_subset = coms[start_idx:end_idx]
    means_subset = list_of_means[start_idx:end_idx]
    
    sns.barplot(x=coms_subset, y=means_subset, ax=axes[i])
    axes[i].set_title(f'Per Capita Income - Communities {start_idx + 1} to {end_idx}')
    axes[i].set_xlabel('Community Area')
    axes[i].set_ylabel('Per Capita Income ($)')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
items     = ["Olive Oil", "Milk", "Bread", "Eggs", "Cheese"]
item_prices = [8.75, 5.50, 3.00, 12.00, 18.50]

total = 0
for i, (item, p) in enumerate(zip(items, item_prices), start=1):
    total += p
    label = item + " " if p > 15 else item
    print(f"{i}. {label:<22}{p:>6.2f} EGP")
print(f"TOTAL: {total:.2f} EGP")
```
